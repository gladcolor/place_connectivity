# AgglomerativeClustering

```python
import pandas as pd
import sklearn
from sklearn.cluster import AgglomerativeClustering
import numpy as np
from natsort import natsorted

import plotly.express as px
import json
from urllib.request import urlopen

with urlopen('https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json') as response:
    counties = json.load(response)
    
def drawmap(df_re):
    fig = px.choropleth_mapbox(df_re, geojson=counties, locations='fips', color='class',
#                                color_discrete_map="Viridis",
#                                color = 'class'
                               range_color=(0, 50),
                               mapbox_style="carto-positron",
                               zoom=3, center = {"lat": 37.0902, "lon": -95.7129},
                               opacity=0.5,
                               labels={'unemp':'unemployment rate'}
                              )
    fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
    fig.show()   
```

```python
gby_list = list(gby_set)
gby_list = natsorted(gby_list)
len(gby_list)
```

## build distance matrix

```python
dis_mat = np.zeros((len(gby_list[:]), len(gby_list[:])))
print("dis_mat shape:", dis_mat.shape)  

fips_dic = {}                
for i in range(len(gby_list[:])):
#     print(i, gby_list[i])
    fips_dic[gby_list[i]] = i
# print(fips_dic)
```

```python
fips_dic
```

```python
df
```

```python
for idx, row in df.iterrows():
    interval = 100000
    if idx % interval == 0:
        print(idx)
#     print(row['o_fips'])
    i = fips_dic[row['o_fips']]
    j = fips_dic[row['d_fips']]
    i = int(i)
    j = int(j)
    dis_mat[i, j] = row['ri']
    dis_mat[j, i] = row['ri']
    
print("done")
```

```python
import scipy
saved_file = r'/media/huan/HD8T/Research/place_conn/PCI_2019_s_matrix.csv'
s_dis_mat = scipy.sparse.coo_matrix(dis_mat)
scipy.sparse.save_npz(saved_file, s_dis_mat, compressed=True)
```

## with connectivity, linkage = 'average'

```python
import scipy
saved_file = r'/media/huan/HD8T/Research/place_conn/PCI_2019_s_matrix.csv.npz'
s_dis_mat = scipy.sparse.load_npz(saved_file)
# s_dis_mat = scipy.sparse.coo_matrix(dis_mat)
dis_mat = s_dis_mat.todense()
dis_mat
```

```python
dis_mat.sum()
```

```python
df['ri'].sum()
```

```python
dis_mat_inv = 1 / (dis_mat/1000 + 0.0000001)

dis_mat_inv

dis_mat_inv_small = dis_mat_inv[:, :]
conn_mat = np.where(dis_mat_inv_small > 99999.9, 0, 1)
print(conn_mat.shape)
conn_mat
```

```python
conn_mat.sum()
```

```python

# n_clusters = 48

df_re_list = []

for n_clusters in (20, 48, 50, 75):
    clustering = AgglomerativeClustering(n_clusters=n_clusters,  affinity="precomputed", linkage='single', \

                                       connectivity=conn_mat).fit(dis_mat_inv_small)


    fips_col = []
    for idx, label in enumerate(clustering.labels_):
    #     print(gby_list[idx], label)
        fips_col.append(gby_list[idx])

    df_re = pd.DataFrame(list(zip(fips_col, clustering.labels_)), columns=['fips', 'class'])

    df_re['fips'] = df_re['fips'].astype(str).str.zfill(5)
    df_re['class'] = df_re['class'].astype(str)


    df_re.to_csv(f'/media/huan/HD8T/Research/place_conn/agglomerative_{n_clusters}cluster_single.csv', index=False)
    df_re_list.append((n_clusters, df_re))

    print("finished ", n_clusters)

```
